# **BOX PLOT**

***This is to show the visualizatio of the result above using Boxplot. We can see the represenation of the mean, 25%,50%, 75% and outliers.***

```python
# Boxplot for detecting outliers
plt.figure(figsize=(15, 5))
for i, column in enumerate(['TV', 'radio', 'newspaper', 'sales'], 1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

# **Histogram**

# ***Key Insight***


*   TV and radio advertising see a more even distribution in spending, possibly due to their effectiveness or the nature of campaigns.

*   Newspaper advertising is much less common for high spending, indicating a lower priority.

*   Sales distribution suggests most companies experience moderate success, with high sales being rare.




```python

# Plot histograms with KDE for each advertising type and sales
plt.figure(figsize=(10, 8))
for i, column in enumerate(['TV', 'radio', 'newspaper', 'sales'], 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

# **Correlation Matrix**

Correlation coefficients measure the strength and direction of the linear relationship between two variables, ranging from -1 (perfect negative correlation) to +1 (perfect positive correlation). Key insight to what this values indicates in relation to sales:



*   TV advertising has the highest correlation with sales **(0.78),** suggesting it's the most effective channel for increasing sales among the three.

*   Radio shows a moderate correlation **(0.58),** implying it still positively affects sales but to a lesser extent than TV.

*   Newspaper has the weakest correlation with sales **(0.23),** indicating it might not be as impactful for driving sales directly, or its effect may vary based on the audience or other factors not captured in this dataset.


These correlation values help prioritize advertising channels by their likely impact on sales.




```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Advertising vs Sales Dataset")
plt.show()
```
